```python
import pandas as pd
import numpy as np

# Importing plotly express which will be used for creating the visualizations
import plotly.offline as py
from plotly.offline import iplot
import plotly.graph_objs as go

#matplotlib library
import matplotlib.pyplot as plt
```

```python
data_path = 'final_result.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
import plotly.express as px

# 标准差方法去除异常值
threshold = 3  # 阈值为 3 倍标准差
mean = df['精神慰撫金額'].mean()
std = df['精神慰撫金額'].std()
processed_df = df[abs(df['精神慰撫金額'] - mean) < threshold * std]
```

```python
fig = px.histogram(processed_df, x="精神慰撫金額", nbins=200)
fig.show()
```

```python
processed_df_2 = processed_df[processed_df['精神慰撫金額'] < 2e6 ]
fig = px.histogram(processed_df_2, x="精神慰撫金額", nbins=200)
fig.show()
```

```python
df = processed_df_2

# 剔除不是 float 类型的数据
df['醫療費用'] = pd.to_numeric(df['醫療費用'], errors='coerce')
df['精神慰撫金額'] = pd.to_numeric(df['精神慰撫金額'], errors='coerce')
df['薪資收入'] = pd.to_numeric(df['薪資收入'], errors='coerce')

df = df.dropna(subset=['醫療費用'])
df = df.dropna(subset=['精神慰撫金額'])
df = df.dropna(subset=['薪資收入'])

df = df.sort_values(by='醫療費用')
```

```python

import plotly.graph_objects as go
import numpy as np

favor_col = "頭頸部"
on_df = df[df[favor_col] == 1]
off_df = df[df[favor_col] == 0]

fig = go.Figure()
fig.add_trace(go.Histogram(x=np.log(on_df["精神慰撫金額"]), nbinsx=45, name='yes'))
fig.add_trace(go.Histogram(x=np.log(off_df["精神慰撫金額"]), nbinsx=45, name='no'))

# Overlay both histograms
fig.update_layout(
    barmode='overlay',
    title_text=favor_col, # title of plot
    xaxis_title_text='Value', # xaxis label
    yaxis_title_text='Count', # yaxis label
)

# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)
fig.show()
```
